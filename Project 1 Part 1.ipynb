```python
def get_hoods(url):
    p = requests.get(url)
    b = BeautifulSoup(p.content, "html.parser")
    column_names = ['Name',"Link"]
    l = []
    links = b.find("ul", class_="index").find_all("a")
    for i in links:
        Link = 'http://www.city-data.com'+i.get('href')
        Name = i.text.strip()
        if 'Des' in Name and 'Moines' in Name:
            d = dict(zip(column_names,[Name,Link]))
            l.append(d)
    return l

#get_hoods('http://www.city-data.com/indexes/neighborhoods/IA/1/')
```

```python
url = 'http://www.city-data.com/indexes/neighborhoods/IA/%d/'
l = []
for page in range(1,14,1):
    print(url%page)
    l = l + get_hoods(url%page)
    time.sleep(2)
```

```python
l_list = []
for d in l:
    l_list.append((d['Link']))
```

```python
import pandas as pd
df = pd.DataFrame(l) # Make a dataframe
df.set_index('Name', inplace = True)
df.to_csv('Urls.csv')
df.head()
```

```python
def get_content(url):
    p = requests.get(url)
    b = BeautifulSoup(p.content, "html.parser")
    datacontent = []
    column_names = ['area','population','mf_pop','avg_size','cars_per_house','cars_per_apartment','median_age_males','median_females','median_income','median_rent','percent_mortgage','percent_born_in _state','percent_born_in_other_state','percent_foreign_born','percent_married','never_married_males_over_15','never_married_females_over_15','percent_do_not_speak_english_well','percent_single_mothers']
    #rows = b.find_all(class_ = 'content-item')
    rows = b.find_all(class_ = 'content-item')
    x = 0
    raw = []
    for r in rows:
        N_Details = b.find_all(class_ = 'content-item')[x].text.strip().splitlines()
        raw.append(N_Details)
        x = x + 1
        
    #content = b.find_all(class_ = 'content-item')
    try:
        area = raw[0][0]
    except:
        area = None

    try:
        population = raw[0][1]
    except:
        population = None

    try:
        mf_pop = raw[4][0]
    except:
        mf_pop = None
        
    try:
        avg_size = raw[5][0]
    except:
        avg_size = None
        
    try:
        cars_house = raw[13][1]
    except:
        cars_house = None
        
    try:
        cars_apart = raw[13][4]
    except:
        cars_apart = None
    
    try:
        median_males = raw[4][1]
    except:
        median_males = None
        
    try:
        median_females = raw[4][2]
    except:
        median_females = None
        
    try:
        median_income = raw[1][1]
    except:
        median_income = None
        
    try:
        median_rent = raw[2][1]
    except:
        median_rent = None
        
    try:
        percent_mortgage = raw[14][1]
    except:
        percent_mortgage = None
        
    try:
        born_state = raw[9][1]
    except:
        born_state = None
        
    try:
        born_us = raw[9][4]
    except:
        born_us = None
        
    try:
        born_out = raw[9][7]
    except:
        born_out = None
        
    try:
        percent_married = raw[5][4]
    except:
        percent_married = None
    
    try:
        never_married_males = raw[7][1]
    except:
        never_married_males = None
        
    try:
        never_married_females = raw[7][4]
    except:
        never_married_females = None
        
    try:
        percent_poor_english = raw[8][1]
    except:
        percent_poor_english = None
        
    try:
        single_mothers = raw[5][10]
    except:
        single_mothers = None
        
            
    d = dict(zip(column_names,[area,population,mf_pop,avg_size,cars_house,cars_apart,median_males,median_females,median_income,median_rent,percent_mortgage,born_state,born_us,born_out,percent_married,never_married_males,never_married_females,percent_poor_english,single_mothers]))
    datacontent.append(d)
    return datacontent
```
