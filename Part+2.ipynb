```python
import pandas as pd
import time
```

```python
df = pd.read_csv('NeighborhoodData2.csv')
```

```python
df.head()
```

```python
df.rename(columns = {'percent_born_in _state':'born_in_state',
                     'percent_born_in_other_state':'born_out_state',
                    'avg_size': 'avg_household_size',
                    'median_age_males':'M_median_age',
                    'median_females':'F_median_age',
                    'never_married_females_over_15':'F_never_married',
                    'never_married_males_over_15':'M_never_married',
                    'percent_do_not_speak_english_well':'speak_poor_english',
                    'percent_foreign_born':'foriegn_born',
                    'percent_married':'married',
                    'percent_mortgage':'with_mortgage',
                    'percent_single_mothers':'single_mothers'}, inplace = True)
```

```python
df.area = df.area.str.extract('([\d.]+)', expand = True)
df.avg_household_size = df.avg_household_size.str.extract('([\d.]+)', expand = True)
df.cars_per_apartment = df.cars_per_apartment.str.extract('([\d.]+)', expand = True)
df.cars_per_house = df.cars_per_house.str.extract('([\d.]+)', expand = True)
df.M_median_age = df.M_median_age.str.extract('([\d.]+)', expand = True)
df.F_median_age = df.F_median_age.str.extract('([\d.]+)', expand = True)
df.median_income = df.median_income.str.extract('([\d,]+)', expand = True)
df.median_income = df.median_income.str.replace(',','')
df.median_rent = df.median_rent.str.extract('([\d,]+)', expand = True)
df.median_rent = df.median_rent.str.replace(',','')
df['male_population'] = df.mf_pop.str.extract('([\d,]+)([\w:]+):([\d,]+)', expand = True)[0]
df['female_population'] = df.mf_pop.str.extract('([\d,]+)([\w:]+):([\d,]+)', expand = True)[2]
df.male_population = df.male_population.str.replace(',','')
df.female_population = df.female_population.str.replace(',','')
del df['mf_pop']
df.F_never_married = df.F_never_married.str.extract('([\d.]+)', expand = True)
df.M_never_married = df.M_never_married.str.extract('([\d.]+)', expand = True)
df.born_in_state = df.born_in_state.str.extract('([\d.]+)', expand = True)
df.born_out_state = df.born_out_state.str.extract('([\d.]+)', expand = True)
df.speak_poor_english = df.speak_poor_english.str.extract('([\d.]+)', expand = True)
df.foriegn_born = df.foriegn_born.str.extract('([\d.]+)', expand = True)
df.married = df.married.str.extract('([\d.]+)', expand = True)
df.with_mortgage = df.with_mortgage.str.extract('([\d.]+)', expand = True)
df.single_mothers = df.single_mothers.str.extract('([\d.]+)', expand = True)
df.population = df.population.str.extract('([\d.]+)', expand = True)
```

```python
df['address'] = df.name.str.replace(' neighborhood in ',',')
```

```python
from geopy.geocoders import Nominatim
def get_neigh_coords(inputAddress):
    try:
        geolocator = Nominatim(user_agent = 'my-application')
        location = geolocator.geocode(inputAddress)
        time.sleep(2)
        print('.')
        return location.latitude,location.longitude
    except:
        return None,None
```

```python
df['coords'] = df.address.apply(lambda x:get_neigh_coords(x))
```

```python
df['lat'] = df.coords.apply(lambda x:x[0])
df['lon'] = df.coords.apply(lambda x:x[1])
```
